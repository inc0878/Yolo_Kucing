# membuat tes prediksi dg vidio

```python
import os
from ultralytics import YOLO
import cv2

# Set path untuk video input dan output
VIDEOS_DIR = os.path.join('.', 'videos')

video_path = os.path.join(VIDEOS_DIR, 'kucing.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

# Buka video
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
    exit()

# Ambil frame pertama untuk mendapatkan ukuran dan frame rate video
ret, frame = cap.read()
H, W, _ = frame.shape
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), fps, (W, H))

# Path ke model custom
model_path = 'D:/Projek/Yolo_Kucing/runs/detect/train11/weights/last.pt'

# Load YOLO model
model = YOLO(model_path)

threshold = 0.5

while ret:
    # Inferensi pada frame
    results = model([frame])[0]  # Bungkus frame sebagai batch

    # Proses setiap deteksi
    for box in results.boxes:
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy().astype(int)
        score = box.conf[0].cpu().numpy()
        class_id = int(box.cls[0].cpu().numpy())  # Ubah class_id menjadi integer

        if score > threshold:
            # Gambar kotak deteksi dan teks kelas
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, results.names[class_id].upper(), (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    # Simpan frame yang telah diproses
    out.write(frame)
    ret, frame = cap.read()  # Baca frame berikutnya

# Lepaskan video dan simpan output
cap.release()
out.release()
cv2.destroyAllWindows()
```
